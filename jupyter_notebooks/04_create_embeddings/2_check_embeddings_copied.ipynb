# Notebook overview
Summarizes which embeddings (high and low) were copied or created and prints counts per dataset split.

- Loads the dataset CSVs with embedding information (high and low)
- Checks 'embedding_created' ('embedding_copied') columns and prints value counts per split
- To validate embedding creation and copying steps

# Preperation

### Import

```python
import pandas as pd
from pathlib import Path
```

### Path - df_destination_dir

```python
DF_DESTINATION_PATH = r'/home/stud/jleick/masterArbeitProjekt/data/datasets/created/embedding_copied'
df_destination_dir = Path(DF_DESTINATION_PATH)
```

### Load df - high

```python
high_id_train = pd.read_csv( df_destination_dir/'high_id_train.csv', index_col=0)
high_id_val = pd.read_csv( df_destination_dir/'high_id_val.csv', index_col=0)
high_id_test = pd.read_csv( df_destination_dir/'high_id_test.csv', index_col=0)

high_ood_test = pd.read_csv( df_destination_dir/'high_ood_test.csv', index_col=0)
```

### Load df - low

```python
low_id_test = pd.read_csv( df_destination_dir/ 'low_id_test.csv', index_col=0)

low_ood_test = pd.read_csv( df_destination_dir/ 'low_ood_test.csv', index_col=0)
```
